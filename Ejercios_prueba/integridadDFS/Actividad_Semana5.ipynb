```python
import urllib.request
url = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/cars_clus.csv'
filename = 'cars_clus.csv'
df=urllib.request.urlretrieve(url, filename)
import pandas as pd
pdf = pd.read_csv(filename,na_values = "$null$")
pdf
```

```python
pdf.isnull().sum()
```

# Parte 1

```python
#la funciÃ³n dropna se utiliza para eliminar filas con valores faltantes (NaN). El argumento subset especifica que se deben considerar las columnas 'manufact' y 'price' para buscar valores faltantes. El argumento how='any' significa que si alguna de las columnas especificadas tiene un valor faltante en una fila, esa fila se eliminarÃ¡
print(pdf.shape)
pdf1=pdf.dropna(subset=['manufact','price'],how='any')
print(pdf1.shape)
```

```python
#dropna se usa para eliminar todas las filas que contengan un valor faltante en cualquier columna
pdf2=pdf.dropna()
print(pdf2.shape)
pdf2
```

```python
#La funciÃ³n groupby se utiliza para agrupar los datos en funciÃ³n de los valores Ãºnicos en la columna 'manufact', y luego describe() se aplica a esas agrupaciones para calcular estadÃ­sticas descriptivas como la media, la desviaciÃ³n estÃ¡ndar, el valor mÃ­nimo, el valor mÃ¡ximo y los percentiles para cada grupo
pdf[['manufact','price']].groupby(by='manufact').describe()
```

# Parte 3

```python
# pdf.dtypes se utiliza para mostrar los tipos de datos de cada columna en el DataFrame 
pdf.dtypes
```

```python
# assign en un DataFrame de Pandas para agregar dos nuevas columnas, 'Valor_nuevo' y 'Valor_nuevo2', al DataFrame pdf.Estas columnas se calculan utilizando funciones lambda que multiplican la columna existente 'price' por un factor.
pdf=pdf.assign(Valor_nuevo= lambda x: x['price']*1.05,\
               Valor_nuevo2= lambda x: x['price']*1.10)
pdf
```

```python
# la funciÃ³n loc de Pandas para seleccionar filas en un DataFrame pdf que cumplan con ciertas condiciones.
pdf.loc[(pdf['sales']>10) &(pdf['price']>10),:]
```

```python
# el mÃ©todo query en un DataFrame de Pandas para seleccionar filas que cumplan con ciertas condiciones
pdf.query("sales >10 and price >10")
```
